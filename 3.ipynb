```python
n = int(input("Indique un número: "))
for i in range(11):
    print(n, "x", i, "=", n*i)
```

```python
for i in range(1,11):
    for j in range(11):
        print(i, "x", j, "=", i*j)
    salir=False
    while salir==False:
        tecla = input("Pulse enter para continuar")
        if tecla =="":
            salir=True    
```

```python
n=int(input("Número de múltiplos= "))
x=int(input("Número= "))
suma=0
salir=False
while(salir==False):
    for i in range(0,n+1):
        suma+=x*i
    print("La suma es", suma)
    if x>0:
        salir=True
```
