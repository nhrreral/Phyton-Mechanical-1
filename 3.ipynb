```python
salir = False
while salir==False:
  x1 = float(input("Dato 1="))
  x2 = float(input("Dato 2="))
  x3 = float(input("Dato 3="))
  x4 = float(input("Dato 4="))
  Ma = (x1+x2+x3+x4)/4
  Mg = pow((x1*x2*x3*x4),(1/4))
  print("La media aritmetica es ", Ma)
  print("La media geometrica es ", Mg)
  opcion = input("\t¿Desea salir? Pulse S|s: ")
  if opcion=="S" or opcion=="s":
    salir = True
```

```python
n=int(input("Número = "))  
cuenta=0
suma=0

for i in range(n, 0, -1):
  if n%i == 0:
    print(i, "es divisor de", n)
    cuenta+=1
    suma+=i 
print("La suma es",suma)
print("El número de divisores es",cuenta)
```

```python
n=int(input("Número = "))  
suma=0
for i in range(1, n+1):
    termino=((-1)**(i+1))/i   #pow((-1),(i+1))
    suma=suma+termino
    print(termino)
print("La suma de la seria es", suma)
```
