```python
n = int(input("Indique un número: "))
for i in range(11):
    print(n, "x", i, "=", n*i)
```

```python
for i in range(1,11):
    for j in range(11):
        print(i, "x", j, "=", i*j)
    salir=False
    while salir==False:
        tecla = input("Pulse enter para continuar")
        if tecla =="":
            salir=True    
```
