```python
from math import pi
from math import sqrt
r = float(input("Valor radio: "))
h = float(input("Valor altura: "))
v = pi*r*r*h/3
print("El volumen es", v)
g = sqrt((pow(h,2))+(pow(r,2)))
Ab = pi*(pow(r,2))
Al = pi*r*g
s = Ab+Al
print("La superficie es", s)
```

```python
from math import sqrt
from math import pow
x1 = float(input("Valor x1= "))
x2 = float(input("Valor x2= "))
y1 = float(input("Valor y1= "))
y2 = float(input("Valor y2= "))
z1 = float(input("Valor y1= "))
z2 = float(input("Valor y2= "))

d1 = sqrt (pow((y2-x2),2)+pow((y1-x1),2))
d2 = sqrt (pow((z2-x2),2)+pow((z1-x1),2))
d3 = sqrt (pow((z2-y2),2)+pow((z1-y1),2))

s = (d1+d2+d3)/2
A = (sqrt(s*(s-d1)*(s-d2)*(s-d3)))
print("A =", A)
```
