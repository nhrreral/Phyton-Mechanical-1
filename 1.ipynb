```python
from math import sqrt
from math import pow
x1 = float(input("Valor x1= "))
x2 = float(input("Valor x2= "))
y1 = float(input("Valor y1= "))
y2 = float(input("Valor y2= "))
z1 = float(input("Valor y1= "))
z2 = float(input("Valor y2= "))

d1 = sqrt (pow((y2-x2),2)+pow((y1-x1),2))
d2 = sqrt (pow((z2-x2),2)+pow((z1-x1),2))
d3 = sqrt (pow((z2-y2),2)+pow((z1-y1),2))

s = (d1+d2+d3)/2
A = (sqrt(s*(s-d1)*(s-d2)*(s-d3)))
print("A =", A)
```

```python
# Sean "s, m, h y d" "segundos, minutos, horas y días", 
# y "sr, mr, hr" los segundos, minutos y horas que sobran respectivamente
s = int(input("tiempo total en segundos"))
m = s//60
sr = s%60
h = m//60
mr = m%60 
hr = h%24
d = h//24
print(s, "segundos son", m, "minutos y sobran", sr, "segundos")
print(m, "minutos son", h, "horas y sobran", mr, "minutos")
print(h, "horas son", d, "días y sobran", hr, "horas")
```

```python
p = float(input("Coste total del préstamo en euros"))
n = float(input("Plazo del préstamo en meses"))
i = float(input("Tipo de interés anual"))
P = (p*(i/1200)*(pow(1+(i/1200),n)))/((pow(1+(i/1200),n))-1)
print("El pago mensual es de", P, "euros")
```
