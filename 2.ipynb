```python
x = float(input("Valor de x= "))
y = float(input("Valor de y= "))
z = float(input("Valor de z= "))
vx = float(input("Valor de vx= "))
vy = float(input("Valor de vy= "))
vz = float(input("Valor de vz= "))
a = float(input("Valor de a= "))
b = float(input("Valor de b= "))
c = float(input("Valor de c= "))
d = float(input("Valor de d= "))
prod = vx*a + vy*b + vz*c
ev = x*a+y*b+z*c+d

if prod==0:
    if ev==0:
        print("La recta está contenida en el plano")
    elif ev<0 or ev>0:
        print("La recta y el plano son paralelos")
elif prod<0 or prod>0:
    print("La recta y el plano son secantes")
    t= -(a*x + b*y + c*z + d )/(a*vx+b*vy+c*vz )
    xc=x+vx*t
    yc=y+vy*t
    zc=z+vz*t
    print("El punto de corte es", xc,yc,zc)
```
