```python
from math import sqrt
from math import trunc

num=float(input("Número= "))
raiz = (sqrt(num))
pe = trunc(raiz)

if (pe*pe == num):
    print(num, "es un cuadrado perfecto")

else:
    print(num, "no es un cuadrado perfecto")
```
